# Boxplots

```python
# import required packages
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Feed file path to read_csv. It maybe different for you
d = pd.read_csv('Data/altcoins.csv')

d.head()

# We don't need Not a Number values, so we will replace
# them with 0
d.fillna(0, inplace=True)
```

```python
# Instantiate a figure and axes
fig, ax = plt.subplots()

# Let's analyze the last 30 days
# of bitcoin prices
data = d.tail(30)

# Choose the two columns
# and make a list
l=[data.ETH, data.DASH]

# Time for our bare-bone boxplot
ax.boxplot(l)
```

```python
# Let us  make another boxplot,


fig1, ax1 = plt.subplots(figsize=(12,5))

bp = ax1.boxplot(l, 
                 vert=False, 
                 whis=1.5,
                widths=[0.3, 0.8],
                labels=['ETH','DASH'],
                showmeans=False, 
                meanprops=dict(marker='.'),
                showcaps=True,
                showfliers=True,
                whiskerprops=dict(linestyle='dashed'),
                flierprops=dict(marker='*', markersize=10, markeredgecolor='r'),
                medianprops=dict(linestyle='dotted', linewidth=3))


ax1.set_title('ETH vs. DASH', size=20)

from matplotlib.ticker import StrMethodFormatter

mft = StrMethodFormatter("${x}")

ax1.xaxis.set_major_formatter(mft)
```

```python
# call the 
bp['boxes'][1].set(color='c', 
                   lw=2, 
                   ls='dashed')


fig1
```

```python
# let's make the outlier markers little bigger
bp['fliers'][0].set(markersize=2)
fig1
```

```python
# But before, we'll find out the x and y values of the outliers
outliers = bp['fliers'][1].get_data()


outliers
```

```python
# we need the first tupple
# and the last value in it
do = outliers[0][1]

# Time for annotation
an = ax1.annotate(str(do), 
             xy=(do, 2), 
             xytext=(do-50, 1.5),
            arrowprops=dict(arrowstyle='->'))


# Let's grab the arrow patch
# and make it a little sophisticated
an.arrow_patch.set_connectionstyle('angle, angleA=90')

# and make the fontsize bigger
an.set_size(16)

fig1
```
