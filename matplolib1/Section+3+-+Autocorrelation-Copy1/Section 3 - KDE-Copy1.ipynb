# KDE

Hi and welcome to the class. Today we will make a kde plot. 
KDE stands for Kernel Density Estimator. It sounds fancy. What it does is that it is a smoothening technique and makes inference of population based on finite data. For example, you have randomly selected 100 tennis players and and measured their weights. And then you try making an inference of weight distribution with their probabilities for all tennis players at some confidence level. 
Let's see it in action.

```python
# import required packages
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
# We will use Weed prices dataset
d = pd.read_csv('Data/df_weed.csv')

# There is a State column with repeating
# values, so let's organize it as a category

d = d.astype({'State':'category'})

# quick preview
d.head()
```

Let's infer a price distributino of all Mary Jane with High, Medium and Low qualities, based on the small dataset at our disposal. For KDE plot, we will Pandas, as it supports KDE plotting right off the bat, unlike axes methods, that have no KDE helpers.

```python
# instantiate a figure and axes
fig, ax = plt.subplots(figsize=(12,5))

# plot straight from pandas
d.plot.kde(ax=ax)
```

To make the chart more informative, let's do basic customization and formating

```python
# Add a title to the chart
ax.set_title('Price distribution of Mary Jane', size=22)

# Let's assign $ symbol to the x axis tick labels
from matplotlib.ticker import StrMethodFormatter

mft = StrMethodFormatter("${x}")
ax.xaxis.set_major_formatter(mft)
ax.set_xlabel('Price ranges')

# Let's zoom in between -100 and 600 interval
ax.set_xlim(-100, 600)


# Remove the spines
spines=['bottom','top','left','right']

for s in spines:
    ax.spines[s].set_visible(False)

# Set gridline
ax.grid(True)

# We will change labels on the legend
ax.legend(labels=['High Quality','Medium Quality','Low Quality'])

# pay attention to the order, as they need to match with column orders

# Finally, let's change the figure and axes
# colors
ax.set_facecolor('#cfd8dc')
fig.set_facecolor('#cfd8dc')   

    
fig
```

Though, interpretation is not the focus of this course, we can infer that price distribution for all Mary Jane with Low, Medium and High qualities actually differ.
