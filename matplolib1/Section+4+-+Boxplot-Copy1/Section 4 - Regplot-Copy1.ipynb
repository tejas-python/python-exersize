# Regplot

Welcome to the class. Today we will make regression plot with Seaborn. 
Speaking of Seaborn, it is high level interface to Matplotlib, and is tailored for making statistical plots.

```python
# Let's import required packages
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

As a dataset, we will use slightly modified version of Iris dataset. To be more precise, it is the same dataset excluding Virginica class.

```python
d = pd.read_csv('Data/setosa-versicolor.csv')
d.head()
```

Let's make a scatter plot of Sepal Length and Petal Width, using regplot. But before, let's find out what we can do with regplot. And of course we ask for help.

```python
help(sns.regplot)
```

Take your time to read through the documentation. It is very important, even if you don't learn it by heart.

```python
# So, our first scatter plot with
# a linear fitting  model
sns.regplot('Sepal length','Petal width', 
            data=d, 
           scatter=True, # this is the default option
            fit_reg = True, # turn on/off fitting line
           order = 1, # for polynomial regression
            n_boot=10,
            truncate=True,
            x_jitter = 0,
            y_jitter=0, # should line fill x axis or not
            marker='^',
            scatter_kws=dict(edgecolor='r',lw=2, facecolor='w', s=50),
            line_kws=dict(color='g', linestyle='dashed'),
            ci=0
        
           )

ax = plt.gca()
fig = plt.gcf()
fig.set_size_inches(12, 4)
```

However, if Y is binary, set logistic to true to make a regression line follow a sigdmoid function. For example.

```python
sns.regplot('Sepal length','Class', 
            data=d, 
            logistic=True,
           n_boot=5 # Since Logistic is more
            #compurationally expensive
            # try decreasing bootstrap number
           )

ax = plt.gca()
fig = plt.gcf()
```
