# Jointplot

Welcome to the class. Today, we will make joinplots
If you read the official description, you may assume it is meant for statisticians only. But you don't have to be a statistician, really. It as an interesting plot that helps you dig further into, a scatter plot.

Too much talking I know, let me just show you.

```python
# Let's import required packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
'''
We will use Tips dataset
'''
d = pd.read_csv('Data/tips.csv')

# view the dataframe
d.head()
```

```python
'''
Unlike countplot, barplot, distplot, 
lmplot and others we have covered so far
Jointplot does not return axes object.
Instead, it returns JointPlot object, which comprises of
multiple axes. So let's do it
'''

g = sns.jointplot('total_bill','tip', data=d)

fig = plt.gcf()
```

What you are looking at, is a collection of three axes on a single figure. Let's break it down

```python
fig.axes
```

In case you need to refer to one of the axes without having to think within seaborn framework, you can do the following

```python
ax1 = fig.axes[0]

ax1.set_facecolor('c')
ax1.collections[0].set(color='red', sizes=[10, 20, 30, 40, 60, 100], alpha=0.4)

fig
```

So, now you know, there is nothing special about JointGrid object. Now, let's get back to seaborn and find out what we can do with it.
We'll start off with help function

```python
help(sns.jointplot)
```

```python
g = sns.jointplot('total_bill','tip', 
                  data=d,
                 kind='reg', #scatter, resid, kde, hex
                 space=0,
                 joint_kws=dict(color='#FF9800'),
                 marginal_kws=dict(color='#3F51B5'),
                 annot_kws=dict(facecolor='#00BCD4', fontsize=14))
```
