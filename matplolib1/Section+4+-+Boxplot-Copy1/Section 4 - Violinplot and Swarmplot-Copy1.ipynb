# Violinplot and Swarmplot

```python
# import required packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Feed file path to read_csv
d = pd.read_csv('tips.csv')

# view the dataframe
d.head()
```

```python
# Let's draw a vertical boxplot grouped by a 
# categorical variable:

fig, ax = plt.subplots(figsize=(12,4))

sns.violinplot(x='day', y='total_bill', data=d, ax=ax, hue='smoker', dodge=True)
```

```python
fig, ax = plt.subplots(figsize=(12,5))
sns.violinplot(x='day', 
               y='total_bill', 
               data=d, 
               ax=ax, 
               hue='smoker', 
               dodge=True, # don't let grouped violibplots overlap
               linewidth=3,
               scale='count', #area, count, width,
               inner='box', #box, quartile, stick, point
               split=True, # draw half the violin on each side,
               # it makes comparison of, say, Yes and No easier
              ) 

ax.tick_params(right=True, labelright=True)
```

```python
fig, ax = plt.subplots(figsize=(12,4))

sns.swarmplot(x='day', y='total_bill', data=d, ax=ax, hue='sex', dodge=True)
```

```python
fig, ax = plt.subplots(figsize=(12,4))

# we make violinplot first
sns.violinplot(x='day', y='total_bill', data=d, ax=ax,  hue='sex', dodge=True)

# and overlaying swarmplot
sns.swarmplot(x='day', y='total_bill', 
              data=d, 
              ax=ax, 
              hue='sex', 
              dodge=True,
             palette=dict(Female='pink',Male='yellow'))

# lets' anchor the left lower border of the legend
# to the middle and make the labels fit in a single row
ax.legend(ncol=4, loc='lower left',bbox_to_anchor=[0, 1])
```
