# Rectangle: Properties

```python
# Import required packages as usual
import numpy as np

# Import charting library
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
# Set a random seed state
np.random.seed(14)

# Let's define x and y axis coordinates
x = list(range(1,6))
y = np.linspace(1, 10, 5)


# Instantiate Figure and Axes as usual
fig, ax = plt.subplots()


# Let's make a simple function to draw 5 rectangles
def makeBars(ax, x, y, w):
    # ax is the axes we will pass
    # x is an x coordinate
    # y is a y coordinate
    # w is width of a patch
    for x, y in zip(x, y):
        bar = Rectangle((x, 0), w, y)
        ax.add_patch(bar)

makeBars(ax, x, y, 0.4)

# We need to make sure that min and max vales of X and Ys
# are within axis ranges

ax.autoscale(tight=False)
```

```python
# Let's make our figure a little bigger.
fig.set_size_inches(15,5)
ax.figure
```

```python
ax.set_title("Some bar chart")
ax.set_xlabel('This is x axis')
ax.set_ylabel('This is y axis')

ax.figure
```

```python
# Set opacity to the first bar

ax.patches[0].set_alpha(0.2)
ax.figure
```

```python
'''
Let's loop through the patches and
change their paramters

'''
for r in ax.patches:
    r.set_color('c')
    r.set_edgecolor('g') # make border color green
    r.set_ls('dashed') # make the border dashed
    r.set_lw(3) # make the border width wider
    r.set_hatch('\\')
    r.set_edgecolor('white')
    
    
ax.figure
```

```python
sketch_magnitude = [0, 0.25, 0.5, 0.75, 1]

for r, s in zip(ax.patches, sketch_magnitude):
    r.set_sketch_params(s)
ax.figure
```

```python
# Let's add another set of 5 rectangles

makeBars(ax, x, y, 0.8)

ax.figure
```

```python
rec_id = range(10)
zorder = reversed(rec_id)
          
for r, z in zip(rec_id, zorder):
    ax.patches[r].set_zorder(z)

ax.figure
```

```python
ax.patches[9].set_clip_on(False)
# here we are saying, turn off the clipping from the dashed line
ax.set_xlim(0, 5.4)
ax.set_ylim(0, 9)
# set xaxis boundary to 39
ax.figure
```
