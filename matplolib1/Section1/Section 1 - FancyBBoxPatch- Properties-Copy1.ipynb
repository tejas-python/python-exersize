# FancyBBoxPatch: Properties

```python
# Import required packages as usual
import numpy as np

# Import charting library
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch
```

```python
# Let's seed a random state
# so we can reproduce random numbers
np.random.seed(14)

# We will also define x and y axis coordinates
x= list(range(1,10))
# x is a list of 10 numbers ranging from 1 through 10

y = list(range(1,10))
# Y is a list of 10 numbers


# Instantiate Figure and Axes
# and set figure size immediately
fig, ax = plt.subplots(figsize=(15,5))


# Let's make a simple function to draw 10 rectangles
def makeFancyBars(ax, x, y, w):
    for x, y in zip(x, y):
        fbox = FancyBboxPatch((x, 0), w, y, boxstyle='round, pad=0.2')
        ax.add_patch(fbox)

makeFancyBars(ax, x, y, 0.3)

# We need to make sure that min and max vales of X and Ys
# are within axis ranges

ax.autoscale()
# and we call this after plotting the chart
```

```python
# Let's add all the possible box styles into a list
box_styles = ['circle', 'darrow', 'larrow','rarrow','round', 'round4','roundtooth','sawtooth','square']

# Let's loop through the patches and apply one box style to a every single patch
for p, s in zip(ax.patches, box_styles):
    p.set_boxstyle(s)

fig

# And get ready to see something, weird
```
