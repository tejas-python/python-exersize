# Line2D: Properties

```python

import numpy as np

# Import charting library
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python

np.random.seed(5)
x = np.linspace(1, 40, 10)

y1 = x / (np.random.rand(10)*2)
y2 = x / (np.random.rand(10)*3)


line1 = Line2D(x, y1, label='Son of a Line')
line2 = Line2D(x, y2, label='Another Line')

fig, ax = plt.subplots()


ax.add_line(line1)
ax.add_line(line2)

ax.autoscale()
```

```python
fig.set_size_inches(15,5)
fig
```

```python
# Let's create handlers for the two lines
# in the axess, as we will be referring to them
# soon
l1 = ax.lines[0]
l2 = ax.lines[1]

# how about we change line1 opacity?
l1.set_alpha(0.3)

fig
```

```python
for line in ax.lines:
    # make all the lines red
    line.set_color('red')    
    # make all the lines wider
    line.set_linewidth(2)
    
ax.figure
```

```python
# Let's make our first line
# dashed

l1.set_linestyle('dashed')
fig
```

```python
l2.set_marker('*')
l2.set_markersize(20)
l2.set_markeredgecolor('red')
l2.set_markerfacecolor('white')
l2.set_markeredgewidth(1)
l2.set_markevery(1)
l2.set_sketch_params(0.9)

fig
```

```python
l1.set_clip_on(False)
# here we are saying, turn off the clipping from the dashed line
ax.set_xlim(0, 39)
# set xaxis boundary to 39
fig
```

```python
ax.set_title("Some line chart")
ax.set_xlabel('This is x axis')
ax.set_ylabel('This is y axis')

ax.figure
```

```python

fig.savefig('artist-line.png', dpi=300, transparent=False)

# We will set the dpi, which is points per pixel
# But this is optional. You can use the default dpi
# Simply by omitting it.

# You may also want to make the background transparent, 
# in case you need to put the chart on top of something
```
