# Axis: Tick Formater and Locator

```python
# import required packages
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# instantiate a figure and an axes
fig, ax=plt.subplots()
```

```python
ax.axis([0, 11, 0, 11])
fig
```

If you choose to be more flexible, you can set x and y limits separately.  For example

```python
ax.set_xlim(0, 12)
ax.set_ylim(0, 8)
fig
```

```python
# Create a handler for X axis
x = ax.xaxis


from matplotlib.ticker import AutoLocator, MaxNLocator, LinearLocator, MultipleLocator, FixedLocator, NullLocator

# let's instantiate those logics

# AutoLocator is the default one
auto_ticks = AutoLocator()

# Max ticks, where you can define
# max number of intervals you need
max_ticks = MaxNLocator(5)

# Number of ticks 
lin_ticks = LinearLocator(5)

# We can also set the interval between ticks
# for example, every 1st, 2nd integer
mul_ticks = MultipleLocator(2)

# OR we can fix the ticks manually
fix_ticks = FixedLocator(locs=[0, 1,4,12])

# We can also remove the ticks
# It has the same affect as hiding
# tick lines and tick labels
null_ticks = NullLocator()
```

```python
# Let's try the autolocator
# We shouldn't see any difference
x.set_major_locator(auto_ticks)
fig
```

```python
# Let's try fitting max number of ticks
# that wouldn't clutter the axis much
x.set_major_locator(max_ticks)
fig
```

```python
# Let's apply linearlocator
x.set_major_locator(lin_ticks)
fig
```

```python
# Time for multiplelocator
x.set_major_locator(mul_ticks)
fig
```

```python
x.set_major_locator(fix_ticks)
fig
```

```python
ax.set_xticks([0,1,2,3,4,12])
fig
```

```python
x.set_major_locator(null_ticks)
fig
```

```python
# Alright, let's reset it back to auto
x.set_major_locator(auto_ticks)
fig
```

```python
from matplotlib.ticker import NullFormatter, FixedFormatter, StrMethodFormatter, ScalarFormatter, PercentFormatter
from matplotlib.ticker import IndexFormatter
```

```python
# Let's instantiate each one

# Null formatter sets null values to
# tick lines
null_form = NullFormatter()

# With FormatStrFormatter you can format the
# the tick labels as strings
str_form = StrMethodFormatter("${x:.2f}")

# ScalarFormatter tick labels as numbers, 
# it is the default option
scala_form = ScalarFormatter()

# Percentformatter turns tick labels into
# percentages
per_form = PercentFormatter(xmax=12)

ind_form = IndexFormatter(list('abcdefghijklmnop'))
```

```python
# Start with nullformatter
x.set_major_formatter(null_form)
fig
```

```python
# OUr string like formatting method
x.set_major_formatter(str_form)
fig
```

```python
# our percentageformatter
x.set_major_formatter(per_form)
fig
```

```python
# and finally our default Formatter
x.set_major_formatter(scala_form)
fig
```

Finally, lets apply index Formatter. It matches list values with scalar values of tick labels.

```python
x.set_major_locator(MultipleLocator(1))
x.set_major_formatter(ind_form)

fig
```
