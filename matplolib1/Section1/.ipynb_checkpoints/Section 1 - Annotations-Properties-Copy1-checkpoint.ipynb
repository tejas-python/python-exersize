# Annotations: Properties

Hi and welcome to the class. Today we will learn how to tweak annotation properties. But first, let me repeat what I mentioned in our previous class.
Whenever you add an Annotation Artist to Axes, you are adding 3 things:
1. Text
2. Box wrapping the text
3. FancyArrowPatch, which is basically the arrow itself

Knowing this will give us strong clues on how far we can go with customizations.

```python
# Let's import pacakges
import numpy as np

# import matplotlib charts
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.text import Annotation


# Charting configs
%matplotlib inline
```

Let's add a text to an empty axes with annotation

```python
annot = Annotation("LooK", xy=(0.2, 0.2), 
                   xytext=(0.5, 0.8), 
                  arrowprops=dict(arrowstyle='->'))

fig, ax = plt.subplots()

# add annoation and assign a hanlder 
# so we can refer to it later
txt1 = ax._add_text(annot)
```

Let's add some title and labels to the axes, and perhaps make the figure little bigger.

```python
fig.set_size_inches(15,5)
ax.set_title('Some Line Chart')
ax.set_xlabel('Some timeline')
ax.set_ylabel('Some price values')
fig
```

Let's explore what options are available for Annotation artist to understand far we can go with customizing it. 

```python
# Let's open the documentation
help(annot)

'''
Please read the documentation
You don't need to memorize it
'''
```

Scroll down to arrowprops. Make sure, you read this part carefully to avoid any confusion... What is actually says, is that If `arrowprops` does not contain the key ``'arrowstyle'`` the allowed keys are:

Alright, time to customize anonation artist. Let's target text and the box wrapping it.

```python
# Recall the FancyBboxPatch properties class
# We will change the style of the  box that wraps the text
txt1.set_bbox(dict(facecolor='red', boxstyle='circle, pad=1.5', ls='dashed', lw=2))
txt1.set_size(14)
txt1.set_color('white')


fig
```

So, we customized the text and box wrapping it. But how can we access the FancyArrowPatch object in Annotation? We simply call arrow_patch attribute like below

```python
arrow = txt1.arrow_patch
arrow
```

If you are curious what parameters FancyArrowPatch objects accept, call the help function

```python
help(arrow)
```

Take your time and read through the documentation. Should not take more than 2 mins.
Now that we know FancyArrowPatch methods, we can start using them.

```python
# Set the line width to 2
arrow.set_linewidth(2) 

# Change arrow style
arrow.set_arrowstyle('-[')

# Let's change connection style
arrow.set_connectionstyle('angle, angleA=-90')
fig
```

There are few other connectionstyles. Feel free to explore them in Annotation chapter of Matplotlib documentation, which you will find in matplotlib.org.

Just for fun, let's try few other connectionstyles.

```python
# arc3
arrow.set_connectionstyle('arc3, rad=-0.5')
fig
```

```python
# Let's try bar
arrow.set_connectionstyle('bar, fraction=-0.1')
fig
```
