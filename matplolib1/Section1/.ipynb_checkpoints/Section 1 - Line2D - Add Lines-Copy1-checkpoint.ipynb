# Line2D: Add Lines

Hi and welcome to the class. Today we will learn how to add a line with __Line Artist__. Normally you simply call plot function and plot a line chart. But whenever you do that, Line Artist comes into play in the background. And of course, you really want to know how things work. 

So, let's get started.

```python
# Import numpy
# we will use it to generate sample data
import numpy as np

# import line2d
from matplotlib.lines import Line2D

# Import charting library
import matplotlib.pyplot as plt

# and we want to view the charts inline
%matplotlib inline
```

```python
# Let's seed a random state
# so we can reproduce the random numbers
np.random.seed(15)

# create a list of 10 digits
# evenly distributed between 1 and 40
x = np.linspace(1, 40, 10)

# let's create a list of Y values
# for two lines
y1 = x / (np.random.rand(10)*2)
y2 = x / (np.random.rand(10)*3)


# Time to instantiate Line2D artists
line1 = Line2D(x, y1)
line2 = Line2D(x, y2)

# Instantiate a figure and axes
# by default x and y axis values
# range between 0 and 1
fig, ax = plt.subplots()


ax.add_line(line1)
ax.add_line(line2)
```

You might be stumbled, why we don't see any lines. It is because, the x and y boundaries of this subplot axes ranges between 0 and 1, meanwhile the numbers we generated do not fit into the scale. Hence we may call a helper function. Autoscale to fix it

```python
ax.autoscale()
fig

# it basically sets ylim and xlim to
# a range that well accomodates the data points
```

Now let's see if we can refer to all or one of the lines on this cart
To do that, we need to call the container that holds reference to Line2D artists


```python
ax.lines
```

As you can see, lines container returns two Line2D objects. Because we have two lines on the charting area, or axes as it is called. Since we know how to call lines container, we can address them easily. So, let's do it

```python
# We will get a little ahead of us and add a color to the line
# But don't worry, we have a special class where we will visit
# line properties
ax.lines[0].set_color('r')

ax.figure
```
