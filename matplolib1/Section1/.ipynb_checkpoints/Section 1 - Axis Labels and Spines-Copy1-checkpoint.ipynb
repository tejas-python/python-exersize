# Axis Labels and Spines

Hi and welcome to the class. Today we will talk about Spines, XAxis and YAxis labels and how one can customize them

```python
# Let's import required packages
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# We will instantiate a figure and two axes
fig, ax = plt.subplots(figsize=(15,5))
```

```python
'''

Let's quickly take a look at the constituents of the
axes

'''
ax.get_children()
```

Here we can quickly spot the objects we are interested in this class, and those are:
1. 4 spines, which are bounding borders 
2. 3 empty text objects, which are Title, X axis label and Y axis label

Let's start our experiment with spines

```python
# call the spines
ax.spines
```

As you can see, it returns a dictionary. Let us now explore parameters of a single spine. We will call Help function for that.

```python
help(ax.spines['left'])
```

As you read through the documentation, you will find plethor of parameters you can play with. So, let's try a few of them.

```python
# Hide border or spine on the top 
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Make the bottom border red line
ax.spines['bottom'].set_color('r')

# Make the right boder thicker
ax.spines['left'].set_lw(2)
ax.spines['left'].set_ls('dotted')
ax.figure
```

Now that we know how spines can be customized, let's move on to Labels.

We will call our beloved function help to explore possible methods and attributes pertaining to axis label.

```python
help(ax.xaxis.get_label())
```

It is a pretty long list again. But let's do the things we assume very obvious.

```python
# Lets give x axes a label
ax.xaxis.set_label_text('I am X')
# alternatively, there is a shortcut
# ax.set_xlabel('I am X')

# Get the label instance, not a text
x_label = ax.xaxis.get_label() # or ax.xaxis.label



# Let's make the label little bigger
x_label.set_size(20)
x_label.set_fontweight('bold')
x_label.set_color('white')



# Since this is a Text Artist
# We can access the wrapping box
# like we did in our earlier classes
# and change the styles

x_label.set_bbox(dict(facecolor='green', 
                      boxstyle='rarrow, pad=0.2',
                     edgecolor='r',
                     linestyle='dashed',
                     linewidth=2))


# You can also move it along x axis 
# by setting x coordinate
x_label.set_x(0.2)


ax.figure
```

```python
# If you want to add more padding 
# between label and xaxis
# you can change the labelpad attribute value
# or play with artist position on x and y coordinates

# let's try with padding instead
ax.xaxis.labelpad=20

ax.figure
```

```python
# You can also toggle
# the visibility of the legend
# just like you could with any Text
# Artists

x_label.set_visible(True)
# Let's set it back to True

ax.figure
```
