```python
np.random.seed(15)
x= np.linspace(1,40,10)
y1=x/(np.random.rand(10)*2)
y2=x/(np.random.rand(10)*2)

line1 =Line2D(x,y1)
line2=Line2D(x,y2)
fig,ax=plt.subplots()
ax.add_line(line1)
ax.add_line(line2)
```

```python
np.random.seed(15)
x= np.linspace(1,40,10)
y1=x/(np.random.rand(10)*2)
y2=x/(np.random.rand(10)*2)

line1 =Line2D(x,y1,label='son of a line')
line2=Line2D(x,y2,label='Another line')
fig,ax=plt.subplots()
ax.add_line(line1,)
ax.add_line(line2)
ax.legend()
ax.autoscale()
```
