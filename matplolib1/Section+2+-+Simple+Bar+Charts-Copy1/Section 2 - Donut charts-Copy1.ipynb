# Donut charts

```python
# import required packages
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Feed file path to read_csv
df = pd.read_csv('Data/barchart-divorcerates.csv')
df.head()
```

```python
fig, ax = plt.subplots()

df.plot.pie(y='Women', labels=df.Age, autopct='%1.1f%%', ax=ax)
            
ax.set_aspect('equal')
ax.legend_.set_visible(False)
ax.set_ylabel('')
```

Alright. We have a pie chart. Now, we need to create an empty circle artist and cast onto the pie chart. And we do it as follows

```python
from matplotlib.patches import Circle

circle = Circle((0,0), 0.75, facecolor='white')

# Once we have the empty circle with a white facecolor
# we just need to add it to the same axes

ax.add_artist(circle)
fig
```
