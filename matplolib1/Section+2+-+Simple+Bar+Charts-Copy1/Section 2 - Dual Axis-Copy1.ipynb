# Dual Axis Charts

```python
# import required packages
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Pass file path to read_csv
d = pd.read_csv('Data/sales.csv')

# quick glance
d
```

```python
fig, ax1 = plt.subplots(figsize=(12,5))

ax1.bar(d.index, d.Quantity, label='Quantity')
t = ax1.set_xticks(d.index)
tl = ax1.set_xticklabels(d.Date)
```

```python
ax1.set_facecolor('#8E24AA')
fig.set_facecolor('#8E24AA')
ax1.set_title('Sales by Month', size=22, color='w')
ax1.tick_params(colors='#FFEB3B', bottom='Off', left='Off', axis='y', labelsize=13)

for p in ax1.patches:
    p.set_facecolor('#FFEB3B')
    an = ax1.annotate(str(p.get_height()), (p.get_x(), p.get_height()*1.01))
    an.set(color='w', size=12)
    

spines=['bottom','top','left','right']

for s in spines:
    ax1.spines[s].set_visible(False)
    
    
from matplotlib.ticker import StrMethodFormatter

mft = StrMethodFormatter("{x:.0f} pc.")

'''
If you need to know more about formatting options
check out the link https://pyformat.info/
'''

ax1.yaxis.set_major_formatter(mft)


fig
```

```python
ax2 = ax1.twinx()
'''
This means, the two charts will have a common
x axis. 
It will also automatically put the labels and ticks
on the opposite side, providing a better look
'''

ax2.plot(d.index, d.Amount, color='#FF5722')


fig
```

```python
for s in ax2.spines:
    ax2.spines[s].set_visible(False)
    
fig
```

```python
ax2.tick_params(colors='w', labelsize=13)

mft2 = StrMethodFormatter("${x:,.2f}")
ax2.yaxis.set_major_formatter(mft2)
ax1.tick_params(axis='x', colors='w', labelsize=14)


fig
```

```python
from matplotlib.lines import Line2D

# legend for Amount
lam = Line2D([],[], color='#FF5722', label='Amount')
```

```python
ax1.legend(handles=[ax1.patches[0], lam], 
           labels=[ax1.containers[0].get_label(), lam.get_label()],
          loc='lower center', ncol=2)

ax1.legend_.set_bbox_to_anchor([0.5, -0.2])

fig
```

Alright. let's try making a dual axis chart with Pandas

```python
fig, ax = plt.subplots(figsize=(12, 4))

d.plot.bar(y='Quantity', label='Quantity', ax=ax, color='y')
d.plot.line(y='Amount', label='Amount', ax=ax, secondary_y=True, mark_right=False)
labels = ax.set_xticklabels(d.Date)
```
