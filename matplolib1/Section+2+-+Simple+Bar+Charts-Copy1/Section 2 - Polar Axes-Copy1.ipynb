# Polar Axis

Hi and welcome to the class. Up until now, we were playing with simple linear axis and today, we will learn how to make polar charts using polar axis in Matplotlib.

I should say, a lot of things are the same. Hence, whatever we learned up until now, all is valid. There are only few extra things we should get to know. Enough of talk, let's just do it.

```python
#import required packages
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# We will use Men and Women Divorce Rate dataset
df = pd.read_csv('Data/barchart-divorcerates.csv')
df.head()
```

```python
# Imagine, we are just creating a stacked bar chart
# It is very easy to do with Pandas
# hence, we will use matplotlib OOP
# just to make sure we stil rememeber the process

fig, ax = plt.subplots(figsize=(12,5), subplot_kw=dict(polar=True))

x = df.Age
y1 = df.Women
y2 = df.Men

# Women on top.. of the stacked bar chart
ax.bar(x, y1+y2, label='Women', alpha=0.8)

# Men below
ax.bar(x, y2, label='Men', alpha=0.8)

# let's play with tick paramters
ax.tick_params(labelleft='Off', left='Off')

l = ax.legend(loc='upper right')
l.set_bbox_to_anchor((0.3, 0.9))
```

If you are still wondering what on earth we are making a stacked bar chart on a linear axis, I can say that we can warp and make it circular.
All we need to do is to pass polar equals True. 

Since, we cannot change the existing axes projection, let's scroll up and add the paramter.

Alright, we just now, converted the linear axis to non-linear axis simply by passing a paramter.  
X axis becomes Theta, or the boundary and Y becomes the radius.

To see what methods and attributes are at our disposable with polar axis, let's call the help function

```python
help(ax)
```

If you need to get a segment of axes. You can pass starting and ending degrees in theta.

```python
ax.set_thetamin(-45)
ax.set_thetamax(90)
fig
```

You can alternatively use x_lim, like we would do with linear axis. The only difference, is that you need to pass a lenth interval. And you know that the length of the periphary of any circle is 2 pi multiplied by r. and Pi equals 3.14.

So, if you want to see half of the circular axis, we will do the following.

```python
# Let's import Pi
# I can manually type 3.14
# But that would not be very precise

import numpy as np
p = np.pi

# set the length interval
ax.set_xbound(0, p)

fig
```
