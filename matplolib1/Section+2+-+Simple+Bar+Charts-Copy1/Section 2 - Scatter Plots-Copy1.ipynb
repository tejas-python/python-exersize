# Scatter plots

```python
# import required packages
import pandas as pd
from pandas import DataFrame

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Feed the file path to read_csv
df = pd.read_csv('Data/index.csv')

# a quick glance
df.head()
```

```python
df = df.astype({'Year':'category', 'Day':'category'})
```

```python
# pick the two variables
a = 'Federal Funds Target Rate'
b = 'Unemployment Rate'

# instantiate a figure and axes
fig, ax = plt.subplots(figsize=(12,5))

p = ax.scatter(a, b, data=df)
```

Great. We have the scatter plot documentation. Let's customize it.

```python
ax.set_title('Unemployment vs. Inflation', size=22)
ax.set_xlabel('Interest Rate')
ax.set_ylabel('Unemployment Rate')

# let's add percent sign to both x and  Ys
# We covered it in our class on Ticks 
# In matplotlib anatomy section
from matplotlib.ticker import StrMethodFormatter

mft = StrMethodFormatter('{x} %')

# Apply the function
ax.xaxis.set_major_formatter(mft)
ax.yaxis.set_major_formatter(mft)

fig
```

```python
# let's add some transparency
p.set_alpha(0.6)

# We can change the size of the markers
# by passing a list. I'll pick Interest Rate
# You can pick or set whatever you like
p.set_sizes(df[a].dropna()**(2.8))

# We can also add alternating colors
p.set_color(['w','y'])

# And let's remove the spines
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# We can also set a nice background colors
# on axes and figure
ax.set_facecolor('#01579b')
fig.set_facecolor('#01579b')


fig
```

```python
# Let's change the title and label colors
ax.title.set_color('White')
ax.tick_params(colors='white')

# Change the y label color
ax.yaxis.get_label().set_color('white')

# you can also simply call label attribute
ax.xaxis.label.set_color('w')
fig
```

Let's try with pandas

```python
fig2, ax2 = plt.subplots(figsize=(12,5))
df.plot.scatter(a, b, ax=ax2)
```

```python
# you can also set it the paramters whle plotting, 
# instead of doing it after

ax2.set_title('Unemployment vs. Interest Rate')
ax2.set_xlabel('Employment Rate ')
ax2.set_ylabel('Interest Rate')

# dataframe.plot.scatter returns axes 
# to reach PathCollection from the axes
# need to call collections container
path_collection = ax2.collections[0]

path_collection.set_sizes(df[a].dropna()**2.8)
path_collection.set_color(['w','y'])
path_collection.set_alpha(0.6)

ax2.title.set_color('w')
ax2.xaxis.label.set_color('w')
ax2.yaxis.label.set_color('w')

ax2.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.spines['right'].set_visible(False)

ax2.set_facecolor('#01579b')
fig2.set_facecolor('#01579b')

ax2.tick_params(colors='w')

fig2
```
