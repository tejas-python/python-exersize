Let's also make a pie chart, straight from pandas

```python
# instantiate figure and axes
fig2, ax2 = plt.subplots()


df.plot.pie(y='Women', labels=df.Age, autopct='%1.1f%%', explode=[0,0.1, 0.1, 0, 0], ax=ax2)
ax2.set_aspect('equal')
ax2.set_ylabel('')
ax2.legend_.set_visible(False)
ax2.set_title('Divorce Rate of Women by Age Group', size=20)


colors=['#FFC107','#FF5722','#A5D6A7','#03A9F4','#CDDC39']
fig2.set_facecolor('#3F51B5')
for t in ax2.texts:
    t.set_color('w')
    t.set_size(11)

for w, c in zip(ax2.patches, colors):
    w.set_facecolor(c)
    w.set_ls('dotted')
    w.set_edgecolor('w')
    w.set_lw(2)
    
ax2.title.set_color('w')
```

You can also pass wedge and text properties when initializing a Pie chart, regardless you are plotting with or without pandas. 
For example

```python
fig3, ax3 = plt.subplots()

df.plot.pie(y='Women', labels=df.Age, autopct='%1.1f%%',
            explode=[0, 0.1, 0.1, 0, 0],
            wedgeprops=dict(ls='dotted', ec='white', lw=2), 
            textprops=dict(color='w', fontsize=12),
            ax=ax3)
ax3.set_aspect('equal')

ax3.legend_.set_visible(False)
ax3.set_ylabel('')
fig3.set_facecolor('#3F51B5')
```
