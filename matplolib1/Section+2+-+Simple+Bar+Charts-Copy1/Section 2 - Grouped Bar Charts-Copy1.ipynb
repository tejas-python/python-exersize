# Grouped Bar Charts

```python
# import required packages
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Feed proper file path. It maybe different for you
df = pd.read_csv('Data/barcharts-us-music-sales-by-genre.csv', index_col='Year')

# quick glance
df.head()
```

```python
# instantiate a figure and an axes
fig, ax = plt.subplots(figsize=(15,5))

# We want to fix the width of the bars
width=0.3


rock = ax.bar(df.index, 'Rock', data=df, label='Rock %', width=width)
jazz = ax.bar(df.index+width, 'Jazz', data=df, label='Jazz %', width=width)


# Let's show the 
ax.legend()
```

Pay attention to the tick locations and tick labels

```python
# We want all the years or x tick labels
# to show up and 
ax.set_xticks(df.index +width/2)
# and add half of its width  to 
# move ticks to the middle of the two bars

# However, the tick labels look weird now
# So, let's set the ticks manually
ax.set_xticklabels(df.index)


fig
```

```python
fig2, ax2 = plt.subplots(figsize=(15,5))


# Choose the columns we want to plot
# and make it plot on ax2
df[['Rock','Jazz']].plot.bar(ax=ax2)
```

```python
# Set a title to the axes
ax2.set_title('This is from Pandas', size=22)

# Let's remove rotation of tick labels
ax2.tick_params(labelrotation=0)

# Finally, let's add some values on 
# top of each bar
for p in ax2.patches:
    an = ax2.annotate(str(p.get_height())+'%', (p.get_x(), p.get_height()))
    an.set_size(9)
fig2
```

Let's try with pandas

```python
# You can make plots without instantiating
# figure and axes, as it does implicitly
ax1, ax2 = df[['Jazz','Rock']].plot.bar(subplots=True)
# now, if you need a handler for a figure
# you can get the current figure  doing the following
fig3 = plt.gcf()

# We can now set a figure size
fig3.set_size_inches(15,5)
```

And you can easily customize any of the multiple plots like below.

```python
ax1.set_title('I am a new title')
ax1.legend_.set_visible(False)
fig3
```
