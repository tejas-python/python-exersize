# Simple Bar Charts

Hi and welcome to the class. Today we will make bar charts. That simple.

```python
# import required packages
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

We will use a dataset on US Music Sales by Genre. You can download the csv the file in the resource section of this class. And
please do so.

```python
# Here is our dataset
df = pd.read_csv('Data/barcharts-us-music-sales-by-genre.csv', index_col='Year')

# quick glance
df.head()
```

So, let's make a bar chart Jazz vs. Rock to compare their shares in the US market.

```python
fig, ax = plt.subplots(figsize=(15,5))

rock = ax.bar(df.index, 'Rock', data=df, label='Rock %')
jazz = ax.bar(df.index, 'Jazz', data=df, label='Jazz %')

# Let's show the legend
ax.legend()
```

Now, we will introduce the absolute necessities to make it more informative.

```python
# add a title to the chart
ax.set_title('Rock vs. Jazz', size=22)

ax.set_xticks(df.index)

# Let's remove the spines
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.legend(frameon=False)

# or you can use tick_params
ax.tick_params(left='Off', right='Off', bottom='Off', top='Off')
fig
```

Let's also add the values on top of each bar. That's loop through the patches of ROck, get their heights and add annotation

```python
for p in rock.patches:
    t = ax.annotate(str(p.get_height())+'%', (p.get_x(), p.get_height()+2))
    t.set(color='black', size=8)
    
    t.set_bbox(dict(facecolor='white', boxstyle='circle',lw=1, edgecolor='red'))

fig
```
