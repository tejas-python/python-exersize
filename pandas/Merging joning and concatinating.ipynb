### Inner joins in pandas

```python
week1.head()
```

```python

week2.head()
```

```python
week1.merge(week2,how="inner",
            on="Customer ID",suffixes=["_week1","_week2"])
```

### Join on multiple columns

```python
print(week1.head(2))
week2.head(2)
```

```python
week1.merge(week2,how="inner",on=['Customer ID',"Food ID"])
```

### Outer joins


```python
week1.merge(week2,how="outer",on=['Customer ID'])
```

```python
merged = week1.merge(week2,how="outer",on=['Customer ID'],
           indicator=True)
```

```python
merged["_merge"].value_counts()
```

```python
merged[merged['_merge'].isin(['left_only','right_only'])]
```

### Left join 


```python
foods.head(3)
```

```python
week1.merge(foods,how='left',on="Food ID",
           sort=True)
```

### the left on and right on  prameters

```python
customers.tail(3)
```

```python
week2.merge(customers,how='left',left_on="Customer ID",right_on="ID").drop("ID",axis="columns")
```

### merging index using left and right index paramters

```python
customers.set_index("ID",inplace=True)
```

```python
foods.set_index("Food ID",inplace=True)
```

```python
sales = week1.merge(customers,how="left",left_on="Customer ID",right_index=True)
sales.head(2)
```

```python
sales = sales.merge(foods,how='left',left_on="Food ID",right_index=True)
```

```python
week1.merge(week2,how="left",left_index=True,right_index=True)
```
