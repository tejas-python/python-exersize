### A review of the .apply() Method

```python
bond.head(2)
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + "MILLIONS!"
```

```python
bond["Box_Office"].apply(convert_to_string_and_add_millions)
```

```python
columns = ["Box_Office","Budget","Bond_Actor_Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions)
```

```python
bond.head(2)
```

### The apply() method with row values

```python
bond = pd.read_csv("./pandasfiles/jamesbond.csv")
bond.set_index("Film",inplace=True)
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def good_movie(row):
    actor = row[1]
    budget =row[4]
    if actor =="Pierce Brosnan":
        return "the best"
    elif actor =="Roger Moore" and budget >40:
        return "enjoyable"
    else:
        return "I have no clue"
bond.apply(good_movie,axis="columns")
```
