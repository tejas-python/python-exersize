```python
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
```

### use the plot method to render a line chart

```python
bb = data.DataReader(name="BB",data_source="yahoo",start="2007-01-01",end="2020-12-31")
bb.head(3)
```

```python
bb.plot()
```

```python
bb.plot(y="High")
```

```python
bb["High"].plot()
```

```python
bb[["High","Low"]].plot()
```

### Modifing plot Athestetics with matplotlib Templates

```python
plt.style.use("fivethirtyeight")
bb.plot(y="Close")
```

```python
plt.style.use("dark_background")
bb.plot(y="Close")
```

```python
plt.style.use("ggplot")
bb.plot(y="Close")
```

### Creating bar graphs

```python
def rank_per(stock_price):
    if stock_price <=10:
        return "poor"
    elif stock_price <=50:
        return "Satisifactory"
    else:
        return "Stellar"
```

```python
bb['Close'].apply(rank_per).value_counts().plot(kind="barh")
```

### Creating pie charts to represent proportions

```python
bb['Close'].apply(rank_per).value_counts().plot(kind="pie",legend=True)
```
