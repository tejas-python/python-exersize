```python
import matplotlib.pyplot as plt
import math
from matplotlib.ticker import MultipleLocator
from matplotlib.ticker import FormatStrFormatter
from matplotlib.patches import Rectangle
```

```python
with open('goals.txt','r') as goals:
    Hometeamgoasl = list(map(int,goals.readline().strip(" \n").split(" ")))
    awayteamgoasl = list(map(int,goals.readline().strip(" \n").split(" ")))
```

### Take the data from goals txt which contains home team and away team goals .

```python
plt.scatter(Hometeamgoasl,awayteamgoasl)
plt.show()
```

```python
# unique colours
plt.scatter(Hometeamgoasl,awayteamgoasl,s=50,
            c=awayteamgoasl)
plt.show()
```

```python
#vcolour coding from distance form Zero
def distancefromxzero(htg,atg):
    return math.sqrt(htg**2+atg**2)
colours = []
for i in range(0,len(Hometeamgoasl)):
    colours.append(distancefromxzero(Hometeamgoasl[i],awayteamgoasl[i]))
    
plt.scatter(Hometeamgoasl,awayteamgoasl,s=50,
            c=colours)
plt.show()
```

```python
#marker *,^,-
plt.scatter(Hometeamgoasl,awayteamgoasl,s=50,
            c='r',marker='^')
plt.show()
```

```python
# alpha
plt.scatter(Hometeamgoasl,awayteamgoasl,s=50,
            c='r',marker='^',alpha=0.3)
plt.show()
```

```python

# adding major and minor tick formater and string formatter

majorTickLocation =MultipleLocator(10)
minorTickLocation = MultipleLocator(2)
majorTickFormater = FormatStrFormatter("%03d")# eg 1 to 001 , so it matches 100
ymajorTickFormater = FormatStrFormatter("%02d")# eg 1 to 001 , so it matches 100

fig2 =plt.figure("Second",figsize=(10,10))
ax1 = fig2.add_subplot(2,1,1)
ax1.plot(x,Hometeamgoasl,color='pink')
ax1.set_title("home team goals",loc='center')
ax1.set_xlabel("Game number",labelpad=10)# label pad add space from x axis line font dict can be added
ax1.set_ylabel("home team goals",labelpad=10)# label pad add space from x axis line font dict can be added
ax1.xaxis.set_major_locator(majorTickLocation)
ax1.xaxis.set_minor_locator(minorTickLocation)
ax1.xaxis.set_major_formatter(majorTickFormater)
ax1.yaxis.set_major_formatter(ymajorTickFormater)

plt.show()
```

```python
# bar plots
avgGoalsScoredHome = sum(Hometeamgoasl)/len(Hometeamgoasl)
avgGoalsScoredaway = sum(awayteamgoasl)/len(awayteamgoasl)
```

```python
fig2 =plt.figure("Second",figsize=(10,10))
ax1 = fig2.add_subplot(1,1,1)
ax1.bar(x=[1,2],
        height=[avgGoalsScoredHome,avgGoalsScoredaway],
       width=0.25,
       align = "center", color=('r','black'))
plt.xticks([1,2],["home","away"])
ax1.set_ylim(0,2)      

plt.show()
```

```python
fig2 =plt.figure("Second",figsize=(10,10))
ax1 = fig2.add_subplot(1,1,1)
ax1.bar(x=[1,2],
        height=[avgGoalsScoredHome,avgGoalsScoredaway],
       width=0.25,
       align = "center", color=('r','black'),
       bottom=[0.2,0.2])
plt.xticks([1,2],["home","away"])
ax1.set_ylim(0,2)      

plt.show()
```

```python
fig2 =plt.figure("Second",figsize=(10,10))
ax1 = fig2.add_subplot(1,1,1)
ax1.bar(x=[1,1],
        height=[avgGoalsScoredHome,avgGoalsScoredaway],
       width=0.25,
       align = "center", color=('r','black'),
       bottom=[0,avgGoalsScoredHome])
plt.xticks([1,2],["home","away"])
ax1.set_ylim(0,4)      
ax1.set_xlim(0.5,2.5)      

plt.show()
```

```python
# error bar
fig2 =plt.figure("Second",figsize=(10,10))
ax1 = fig2.add_subplot(1,1,1)
ax1.bar(x=[1,1],
        height=[avgGoalsScoredHome,avgGoalsScoredaway],
       width=0.25,
       align = "center", color=('r','black'),
       bottom=[0,avgGoalsScoredHome],
       xerr= [[0.2,0.3],[0.4,0.1]],
       yerr= [[0.1,0.05],[0.1,0.2]],
       ecolor="blue")
plt.xticks([1,2],["home","away"])
ax1.set_ylim(0,4)      
ax1.set_xlim(0.5,2.5)      
      

plt.show()
```

```python
# error bar
fig2 =plt.figure("Second",figsize=(10,10))
ax1 = fig2.add_subplot(1,1,1)
ax1.boxplot(x=[Hometeamgoasl,awayteamgoasl],
#             positions=[0.5,5])     
               sym="o", # outliers represnet style
                whis =5,
             #  widths =0.5)
                notch =True)# confidence interval[[],[]]
ax1.set_ylim(0,7)
plt.show()
```
